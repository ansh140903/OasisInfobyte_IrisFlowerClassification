### Iris Flower Classification


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Loading Dataset


```python
columns = ['Sepal Lenght','Sepal Width','Petal Length','Petal Width','Class-Label']
df = pd.read_csv('Iris.csv',names = columns)
df.head()
```

```python
df.describe()
```

```python
df['Sepal Lenght'].dtype
df['Sepal Lenght'] = pd.to_numeric(df['Sepal Lenght'])
sns.pairplot(df, vars=["Sepal Lenght", "Sepal Width", "Petal Length", "Petal Width"], hue="Class-Label")
```

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(y_train)
```

Model 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
prediction = model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction))
# for i in range(len(prediction)):
  # print(y_test[i],prediction[i])
```
